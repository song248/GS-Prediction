```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from skimage.morphology import skeletonize
```

```python
img = cv2.imread('chk.png', cv2.IMREAD_COLOR)
print(img.shape)
plt.imshow(img)
```

```python
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
print(img_gray.shape)
plt.imshow(img_gray)
```

```python
_, binary_img = cv2.threshold(img_gray, 160, 255, cv2.THRESH_BINARY_INV)
print(binary_img.shape)
plt.imshow(binary_img)
```

```python
contours, _ = cv2.findContours(binary_img, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
areas = [cv2.contourArea(contour) for contour in contours]
areas
```

```python
image_color = cv2.cvtColor(binary_img, cv2.COLOR_GRAY2BGR)
cv2.drawContours(image_color, contours, -1, (0, 255, 0), 2)
plt.imshow(cv2.cvtColor(image_color, cv2.COLOR_BGR2RGB))
```
