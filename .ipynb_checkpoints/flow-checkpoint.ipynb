```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('for_test.png', cv2.IMREAD_COLOR)
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
print(img_rgb.shape)
plt.imshow(img_rgb)
```

```python
gray_image = cv2.cvtColor(img_rgb, cv2.COLOR_BGR2GRAY)
plt.imshow(gray_image)
```

```python
resized_image = cv2.resize(gray_image, (1440, 1024))
plt.imshow(resized_image)
```

```python
h, w = resized_image.shape  # 흑백 이미지이므로 채널 없음
center_x, center_y = w // 2, h // 2
cropped_image = resized_image[
    center_y - 256:center_y + 256, center_x - 256:center_x + 256
]
plt.imshow(cropped_image)
```

```python
filtered_image = cv2.bilateralFilter(cropped_image, -1, 10, 10)
plt.imshow(filtered_image)
```

```python
inverted = cv2.bitwise_not(filtered_image)
plt.imshow(inverted)
```

```python
kernel = np.ones((3, 3), np.uint8)
dilated = cv2.dilate(inverted, kernel, iterations=1)
thick_lines = cv2.bitwise_not(dilated)
plt.imshow(thick_lines)
```

```python
cv2.imwrite('output.png', thick_lines)  # PNG로 저장
```
